```python
poverty2020=pd.read_excel("https://github.com/XiaoLiang28/Xiao/blob/main/poverty2020.xlsx?raw=true")
```

```python
poverty2020.rename(columns={'State': 'State Name'},inplace=True)
```

```python
poverty2020[['State Name', 'Percent']]
```

```python
edu2020['State Name']=edu2020['State Name'].str.strip()
```

```python
poverty2020['State Name']=poverty2020['State Name'].str.strip()
```

```python
mergededu_poverty = pd.merge(edu2020, poverty2020, how='outer', on=['State Name'],indicator=True)
mergededu_poverty
```

```python
mergededu_poverty.drop(mergededu_poverty.tail(2).index,inplace=True) 
```

```python
mergededu_poverty
```

```python
mergededu_poverty.iloc[:,0:3]
```

```python
mergededu_poverty.rename(columns={'State Name': 'State_Name','Percent_x': 'Percent_edu','Percent_y': 'Percent_poverty'},inplace=True)
```

```python
mergededu_poverty
```

```python
mergededu_poverty.iloc[:,0:3]
```

```python
mergededu_poverty.drop(mergededu_poverty.tail(1).index,inplace=True)
```

```python
mergededu_poverty
```

```python
mergededu_poverty.iloc[:,0:3]
```

```python
mergededu_poverty.Percent_edu.plot(kind ="line", color="g",alpha=0.8, grid="True",linestyle="-",animated=True)
mergededu_poverty.Percent_poverty.plot(kind ="line", color="red", alpha = 0.6, grid = True, linestyle="-",animated=True)

plt.legend(loc="best")
plt.xlabel("Percent_edu")
plt.ylabel("Percent_poverty")
plt.title("Percent_edu-Percent_poverty")
plt.show()
```

```python
mergededu_poverty.plot(kind="scatter", x="Percent_edu", y="Percent_poverty", color="green",alpha=0.8)
plt.legend(loc="best")
plt.xlabel("Percent_edu")
plt.ylabel("Percent_poverty")
plt.title("Percent_edu-Percent_poverty")
plt.show()
```

```python
del price2020['State rank']
```

```python
price2020.rename(columns={'State or territory': 'State_Name','Median home price in US$': 'Home_price'},inplace=True)
```

```python
price2020
```

```python
price2020['State_Name']=price2020['State_Name'].str.strip()
```

```python
poverty2020['State_Name']=price2020['State_Name'].str.strip()
```

```python
mergedprice_poverty = pd.merge(price2020, poverty2020, how='outer', on=['State_Name'],indicator=True)
mergedprice_poverty
```

```python
mergedprice_poverty.drop(mergedprice_poverty.tail(3).index,inplace=True)
```

```python
mergedprice_poverty.iloc[:,0:3]
```

```python
mergedprice_poverty.Home_price.plot(kind ="line", color="g",alpha=0.8, grid="True",linestyle="-",animated=True)
mergedprice_poverty.Percent.plot(kind ="line", color="red", alpha = 0.6, grid = True, linestyle="-",animated=True)

plt.legend(loc="best")
plt.xlabel("Home_price")
plt.ylabel("Percent")
plt.title("Home_price-Percent")
plt.show()
```

```python
mergedprice_poverty.plot(kind="scatter", x="Home_price", y="Percent", color="green",alpha=0.8)
plt.legend(loc="best")
plt.xlabel("Home_price")
plt.ylabel("Percent")
plt.title("Home_price-Percent")
plt.show()
```
